```python
import threading

def fun(num):
    for i in range((num)*100, (num + 1)*100):
        try:
            if (df1['structure'][i] == None):
                df1['struc'][i] = [[i, AFLOWDataRetrieval().get_relaxed_structure(df1['aurl'][i])]]
        except Exception as e:
            print("error: ", i, e)

threads = []
for j in range(92):
    x = threading.Thread(target=fun, args=(j))
    threads.append(x)
    x.start()
```
